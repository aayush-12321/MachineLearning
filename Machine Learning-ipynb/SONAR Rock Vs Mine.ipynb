# Sonar Rock vs Mine Classification
This project aims to classify objects as either rock or mine using the Sonar dataset. The dataset contains sonar signal readings bounced off different objects, where each instance has 60 numerical attributes representing energy measurements at various frequencies. The goal is to build a Machine Learning model that can accurately differentiate between rocks and mines based on these sonar signal patterns.

**Project Workflow:**
- Load and explore the dataset

- Preprocess the data (handle missing values, normalize features)

- Visualize patterns and correlations

- Train and evaluate Machine Learning models (Logistic Regression)

- Optimize and compare model performance

By the end of this notebook, we aim to achieve a robust model that can classify sonar signals.

# Importiong the Dependencies


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data Preprocessing

```python
sonar_data.groupby(60).mean()
```

```python
# Seperating data and label
X, y = sonar_data.iloc[:, :-1], sonar_data.iloc[:, -1]
```

```python
print(X)
print(y)
```

## Train and Test Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model Training --> Logistic Regression

```python
print(X_train)
print(y_train)
```

```python
model = LogisticRegression()
```

```python
# Training the logistic regression model with training data
model.fit(X_train, y_train)
```

# Model Evaluation

```python
# Accuracy on training data
X_train_prediction = model.predict(X_train)
training_test_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
training_test_accuracy
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
training_test_accuracy = accuracy_score(X_test_prediction, y_test)
```
